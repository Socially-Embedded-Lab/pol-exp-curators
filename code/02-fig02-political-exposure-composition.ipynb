```python
#==============================================================================
# 02-fig02-political-exposure-composition
# Purpose: to replicate Figure 2 of the paper, in which we show the composition
#          of political exposure across clusters (i.e., the share of politics curated by
#          different actor types among the different clusters).
# Article: "Who is Curating My Feed? Characterizing Political Exposure of
#          Registered U.S. Voters on Twitter"
# Year:    2022
# Authors: Assaf Shamir, Jennifer Oser, Nir Grinberg
#==============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

rename_detailed_dict = {
    0:"Nonpolitical \n(8.9%)",
    1:"OL Oriented\n(1.7%)",
    2:"Average \nConsumers \n(50.5%)",
    3.1:"Partisans \nLeft \n(2.9%)",
    3.2:"Partisans \nRight \n(11.7%)",
    4:"Media \nOriented\n(15.6%)",
    5:"Media \nOriented++\n(4.0%)",
    6:"Media \nSuperconsumers\n(4.7%)",
}

stacked=True
y_label="% of Political Exposure"
show_percent=True
barlabel_fontsize=16
xticklabels=list(rename_detailed_dict.values())
xtickrotation=0
figsize=(16,16)
add_barlabels=False
fz_big=16
fz_small=14
colors=None

colors = [
"gold",
"khaki",
"indianred",
"lightcoral",
"forestgreen",
"yellowgreen",
"skyblue",
"powderblue",
"peachpuff"
]

data = pd.read_csv("../data/02-fig02-political-exposure-composition.csv", index_col='cluster_detailed')

ax = data.plot(kind="bar", stacked=stacked, figsize=figsize, color=colors)

if add_barlabels:
    for i, container in enumerate(ax.containers):
        ax.bar_label(container, fmt="%.1f", label_type="center", fontsize=barlabel_fontsize)

if xticklabels:
    ax.set_xticklabels(xticklabels, rotation = xtickrotation)
    ax.tick_params(axis='x', which='major', labelsize=fz_big)

ax.tick_params(axis='y', which='major', labelsize=fz_small)
ax.set_ylabel(y_label, fontsize=fz_big)
ax.set_xlabel("")
ax.set_ylim([0, 100])
ax.get_legend().remove()
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0, fontsize=14)
plt.show()
```
